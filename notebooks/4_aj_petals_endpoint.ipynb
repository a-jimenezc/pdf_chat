```python
import asyncio
import json
import websockets

async def generate_text():
    async with websockets.connect('wss://chat.petals.dev/api/v2/generate') as websocket:
        # Send open_inference_session request
        open_session_request = {
            "type": "open_inference_session",
            "max_length": 1024
        }
        await websocket.send(json.dumps(open_session_request))
        
        response = await websocket.recv()
        response_data = json.loads(response)
        
        if response_data.get("ok"):
            print("Inference session opened successfully.")
        else:
            print("Failed to open inference session:", response_data.get("traceback"))
            return
        
        # Send generate request
        generate_request = {
            "type": "generate",
            "inputs": "A cat in French is \"",
            "max_new_tokens": 3
        }
        await websocket.send(json.dumps(generate_request))
        
        while True:
            response = await websocket.recv()
            response_data = json.loads(response)
            
            if response_data.get("ok"):
                outputs = response_data.get("outputs")
                stop = response_data.get("stop")
                print("Generated:", outputs)
                if stop:
                    print("Generation completed.")
                    break
            else:
                print("Failed to generate text:", response_data.get("traceback"))
                break

# Run the async function
asyncio.get_event_loop().run_until_complete(generate_text())
```
